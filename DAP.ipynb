# Q2 Write a Python program to insert an element into a sorted list

```python
arr = sorted(set(eval(input("Enter list of elements"))))
key = int(input("Enter element that needs to be added : "))
print("List of elements : ", arr)

def insert(arr, key):
    index = len(arr)
    for i in range(index):
        if key < arr[i]:
            index = i
            
        return arr[:index] + [key] + arr[index:]

updated_arr = insert(arr, key)
print("The edited list :", sorted(updated_arr))
```

# Q3 Write a python program using object oriented programming to demonstrate encapsulation, overloading and inheritance

```python
class Base:
    def __init__(self):
        self.a = "MCA"
        self._c = "Department"

class Derived(Base):
    def __init__(self):
        super().__init__()
        print("Calling protected member of base class:")
        print(self._c)

obj1 = Base()
print(obj1.a)

obj2 = Derived()
```

```python
class A:
    _x = 5  

class B(A):
    def __init__(self, y):
        self._y = y  

    def __add__(self, other):
        return self._y + other._x 

    def display(self):
        print(f"x = {self._x}, y = {self._y}")

obj1 = B(5)
obj2 = B(6)

obj1.display()
obj2.display()
print("obj2 + obj1 =", obj2 + obj1)
```

# Q5 Implement a python program to demonstrate the following using NumPy
### a)	Array manipulation, Searching, Sorting and splitting.
### b)	broadcasting and Plotting NumPy arrays


```python
import numpy as np
import matplotlib.pyplot as plt

arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([[7, 8, 9], [10, 11, 12]])

concatenated_arr = np.concatenate((arr1, arr2))

reshaped_arr = np.reshape(concatenated_arr, (3, 4))

print("Array manipulation")
print("Orginal array")
print(arr1)
print(arr2)

print("Concatenated array")
print(concatenated_arr)

print("Reshaped array")
print(reshaped_arr)
```

```python
sorted_arr = np.sort(reshaped_arr, axis = 0)

print("\n Sorting : ")
print("Sorted array : ")
print(sorted_arr)
```

```python
split_arrays = np.split(sorted_arr, 3)

print("\n Splitting : ")
print("Split Array")
for i, arr in enumerate(split_arrays):
    print(f"Array {i + 1}: {arr}")
```

```python
broadcasted_arr = reshaped_arr + 10

print("\n Broadcasting : ")
print("Original array : ")
print(reshaped_arr)

print("Broadcast array : ")
print(broadcasted_arr)
```

# Q8 Write a Python program to demonstrate the generation of linear regression models.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

x = np.array([[1,2],[2,4],[3,6],[4,8],[5,10],[6,12]])
y = np.array([2,4,6,8,10,12])

model = LinearRegression()
model.fit(x,y)

print("Coefficent: ",model.coef_)
print("Intercept: ",model.intercept_)

newdata = np.array([[7,14]])
prediction = model.predict(newdata)
print(prediction)

plt.scatter(x[:,0],y,color='blue')
plt.plot(x[:,0],model.predict(x),color='red')
plt.show()
```

# Q10 Write a Python program to demonstrate Time series analysis with Pandas.

```python
import pandas as pd 
import matplotlib.pyplot as plt 

date_rng = pd.date_range(start = '2022-01-01', end = '2022-01-10', freq ='D') 
data = {'Value': [10, 15, 8, 12, 14, 20, 18, 25, 22, 30]} 
df = pd.DataFrame(data, index = date_rng) 

print("Time Series Data:") 
print(df) 
```

```python
plt.figure(figsize = (10, 6)) 
plt.plot(df.index, df[ "Value"], marker = 'o', linestyle = '-', color = 'b') 
plt.title('Time Series Analysis') 
plt.xlabel('Date') 
plt.ylabel('Value') 
plt.grid(True) 
plt.show() 
```

```python
print("\nBasic Time Series Operations:") 
print("Minimum Value:", df['Value'].min()) 
print("Maximum Value:", df['Value'].max()) 
print("Mean Value:", df ['Value'].mean()) 
```

```python
weekly_resampled = df.resample("W").mean() 
print("\nResampled to bleekly Frequency:") 
print(weekly_resampled) 
```

# Q11 Write a Python program to demonstrate Data Visualization using Seaborn

```python
import seaborn as sns 
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips 
```

```python
plt.figure(figsize = (8, 6)) 
sns.regplot(x = "total_bill", y = "tip", data = tips) 
plt.title('Scatter Plot with Regression Line') 
plt.show() 
```

```python
plt.figure(figsize = (8, 5)) 
sns.boxplot(x = "day", y = "total_bill", data = tips) 
plt.title("Box Plot of Total Bill by Day") 
plt.show() 
```

```python
plt.figure(figsize = (10, 6)) 
sns.violinplot(x = "day", y = "total_bill", hue = "sex", data = tips, split = True) 
plt.title("Violin Plot of Total Bill by Day and Gender") 
plt.show() 
```

```python
sns.pairplot(tips, hue = "sex") 
plt.suptitle('Pair Plot of Tips Dataset', y=1.02) 
plt.show() 
```
